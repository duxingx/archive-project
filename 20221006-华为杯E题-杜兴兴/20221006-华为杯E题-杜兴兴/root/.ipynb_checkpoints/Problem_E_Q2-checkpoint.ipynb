```python
# 导入数据

soil_humidity = pd.read_excel("./inputdata/基本数据/附件3、土壤湿度2022—2012年.xls",converters={'月份': '{:0>2}'.format})
soil_evaporativeness = pd.read_excel("./inputdata/基本数据/附件4、土壤蒸发量2012—2022年.xls",converters={'月份': '{:0>2}'.format})

for year in range(2012,2023):
    temp_climate = pd.read_excel("./inputdata/基本数据/附件8、锡林郭勒盟气候2012-2022/{}年.xls".format(year),converters={'月份': '{:0>2}'.format}).copy()
    if year == 2012:
        climates=temp_climate.drop(temp_climate.index,axis=0)
    climates = pd.concat([climates,temp_climate],axis=0)
    
```

```python
# 数据合并
data_merge1 = pd.merge(soil_humidity,soil_evaporativeness,on=["年份","月份"])
data_merge2 = pd.merge(data_merge1,climates,on=["年份","月份"])

data_merge2["日期"] = pd.to_datetime(data_merge2["年份"].astype(str) + data_merge2["月份"],format='%Y%m')

data_merge2 = data_merge2.set_index(data_merge2["日期"])
data_merge2 = data_merge2.drop(["日期","年份","月份","站点号","经度(lon)_x","纬度(lat)_x","经度(lon)_y","纬度(lat)_y","海拔高度(m)","经度","纬度"], axis=1)


# 转换为月度数据
# data_merge2 = data_merge2.resample("M").mean()
# 剔除缺失值大于50%的变量
drop_variables = data_merge2.columns[data_merge2.isna().sum() / len(data_merge2) >0.5]
print(drop_variables)
data_merge3 = data_merge2.drop(drop_variables,axis=1).sort_index()

series = data_merge3.copy()
```

```python
'''acorr_ljungbox(x, lags=None, boxpierce=False)函数检验无自相关
lags为延迟期数，如果为整数，则是包含在内的延迟期数，如果是一个列表或数组，那么所有时滞都包含在列表中最大的时滞中
boxpierce为True时表示除开返回LB统计量还会返回Box和Pierce的Q统计量
返回值：
lbvalue:测试的统计量
pvalue:基于卡方分布的p统计量
bpvalue:((optionsal), float or array) – test statistic for Box-Pierce test
bppvalue:((optional), float or array) – p-value based for Box-Pierce test on chi-square distribution
'''

for vv in range(len(variables)):
    v = variables[vv]
    plotdata = rawdata2[v]
    #划分训练集
    index = plotdata.index[int(len(data_merge2)*0.9)]
    train_data = plotdata[:int(len(data_merge2)*0.9)+1]
    test_data = plotdata[int(len(data_merge2)*0.9):]

    differenced = train_data.diff(1).dropna()

    r,q,p = sm.tsa.acf(differenced.squeeze(), qstat=True)
    data = np.c_[range(1,21), q, p]
    table = pd.DataFrame(data, columns=['lag', "{}Q".format(v), "{}Prob(>Q)".format(v)])
    test_stochastic1 = round(table.set_index('lag'),3)
    if vv==0:
        output = test_stochastic1
    else:
        output = pd.concat([output,test_stochastic1],axis=1)
output.to_csv("./output/纯随机检验结果.csv".format(v),encoding="utf_8_sig")
output
# 从上面的分析结果中可以看到，延迟6阶的p值为<0.05，因此可以拒绝原假设，认为该序列不是白噪声序列。
```

```python
#数据平稳性检测
# 用Dickey-Fuller test检验序列是否平稳
def judge_stationarity(data_sanya_one):
    dftest = ts.adfuller(data_sanya_one)
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    stationarity = "是"
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value 
        if dftest[0] > value:
                stationarity = "否"
#     print(dfoutput)
#     print("是否平稳(是/否): %s" %(stationarity))
    return dfoutput, stationarity


for vv in range(len(variables)):
    v = variables[vv]
    
    plotdata = rawdata2[v]
    #划分训练集
    index = plotdata.index[int(len(data_merge2)*0.9)]
    train_data = plotdata[:int(len(data_merge2)*0.9)+1]
    test_data = plotdata[int(len(data_merge2)*0.9):]
    
#     print("\n",v)
    dfoutput, stationarity = judge_stationarity(train_data)
    
    df = pd.DataFrame(dfoutput,columns=[v]).round(3)
    df.loc[len(df)]=[stationarity]
    
    if vv==0:
        output = df
    else:
        output = pd.concat([output,df],axis=1)
        
output.to_csv("./output/平稳性检验结果.csv".format(v),encoding="utf_8_sig")
output
```

```python
# 定阶
#在statsmodels包里还有更直接的函数：
hqic_order = []
aic_min_order = []
bic_min_order = []
for vv in range(len(variables)):
    v = variables[vv]
    print(v)
    
    plotdata = rawdata2[v]
    #划分训练集
    index = plotdata.index[int(len(data_merge2)*0.9)]
    train_data = plotdata[:int(len(data_merge2)*0.9)+1]
    test_data = plotdata[int(len(data_merge2)*0.9):]
    differenced = train_data.diff(1).dropna()

    order = ts.arma_order_select_ic(differenced,max_ar=11,max_ma=11,ic=['aic', 'bic', 'hqic'])

    orderAIC = pd.melt(order["aic"].reset_index(), id_vars=["index"]).rename(columns = {"index":"AR","variable":"MA","value":"AIC"})
    orderBIC = pd.melt(order["bic"].reset_index(), id_vars=["index"]).rename(columns = {"index":"AR","variable":"MA","value":"BIC"})
    orderHQIC = pd.melt(order["hqic"].reset_index(), id_vars=["index"]).rename(columns = {"index":"AR","variable":"MA","value":"HQIC"})
    orderAIC["order"] = "ARMA(" + orderAIC["AR"].astype(str) +"," + orderBIC["MA"].astype(str) + ")"
    orderBIC["order"] = "ARMA(" + orderAIC["AR"].astype(str) +"," + orderBIC["MA"].astype(str) + ")"
    orderHQIC["order"] = "ARMA(" + orderAIC["AR"].astype(str) +"," + orderBIC["MA"].astype(str) + ")"
    orderAIC.drop(["AR","MA"],axis=1,inplace=True)
    orderBIC.drop(["AR","MA"],axis=1,inplace=True)
    orderHQIC.drop(["AR","MA"],axis=1,inplace=True)

    aic_min_order.append(order.aic_min_order)
    bic_min_order.append(order.bic_min_order)
    hqic_order.append(order.hqic_min_order)
    
order_result = pd.DataFrame({"序列":variables,"AIC":aic_min_order,"BIC":bic_min_order,"HQIC":hqic_order})
order_result.to_csv("./output/定阶详情.csv",encoding="utf_8_sig")
order_result
```

```python
# 构建ARMA模型并导出结果
models = []
for v in variables:
    alldata = rawdata2[v]
    #划分训练集
    index = alldata.index[int(len(data_merge2)*0.9)]
    train_data = alldata[:int(len(data_merge2)*0.9)+1]
    test_data = alldata[int(len(data_merge2)*0.9):]
    
    ar = order_result[order_result["序列"]==v]["HQIC"].tolist()[0][0]
    ma = order_result[order_result["序列"]==v]["HQIC"].tolist()[0][1]

    mod = sm.tsa.ARIMA(alldata,order=(ar, 1, ma))
    print(v,"\n",ar,",",ma)
    
    model = mod.fit()
    models.append(model)
    result_df = model.summary()
    
    temp_table1 = pd.DataFrame(result_df.tables[0])
    temp_table2 = pd.DataFrame(result_df.tables[1])
    temp_table3 = pd.DataFrame(result_df.tables[2])
    temp_table1["变量"]=v
    temp_table2["变量"]=v
    temp_table3["变量"]=v
    if v==variables[0]:
        table1 = temp_table1.drop(temp_table1.index,axis=0)
        table2 = temp_table2.drop(temp_table2.index,axis=0)
        table3 = temp_table3.drop(temp_table3.index,axis=0)

    table1 = pd.concat([table1,temp_table1],axis=0)
    table2 = pd.concat([table2,temp_table2],axis=0)
    table3 = pd.concat([table3,temp_table3],axis=0)
    
with pd.ExcelWriter("./output/构建ARMA模型并导出结果.xls") as writer:
    table1.to_excel(writer,sheet_name="table1")
    table2.to_excel(writer,sheet_name="table2")
    table3.to_excel(writer,sheet_name="table3")
```

```python
# 均值、标准差、四分位数
desc_df = rawdata1.describe().T

# 斜率趋势
def trend(seq):
    idx = np.array(range(len(seq)))
    slope =  linregress(idx, seq.squeeze())[0]
    return slope

trend_df = pd.DataFrame(rawdata1.apply(trend)).rename(columns={0:"趋势斜率"})

# k阶中心矩(2阶)
def moment_k(seq, k=2):
    mean = seq.mean()
    moment_k = np.mean((seq-mean)**k)
    return moment_k

moment_k = pd.DataFrame(rawdata1.apply(moment_k)).rename(columns={0:"2阶中心距"})

# 波峰、波谷
peaks = dict()
troughs = dict()
for i in range(len(rawdata1.columns)):
    peaks[rawdata1.columns[i]] = len(signal.argrelextrema(np.array(rawdata1.iloc[:,i]),np.greater)[0])
    troughs[rawdata1.columns[i]] = len(signal.argrelextrema(np.array(rawdata1.iloc[:,i]),np.less)[0])

peaks_df = pd.DataFrame([troughs]).T
troughs_df = pd.DataFrame([troughs]).T
peak_df = peaks_df.rename(columns={0:"波峰"})
trough_df = troughs_df.rename(columns={0:"波谷"})
```

```python
# 建立模型

#划分训练集(按照时间划分)
res_models = []
fig=plt.figure(figsize=(12,9))
for i in range(len(Y_data.columns)): 

    index = X_newdata.index[int(len(X_newdata)*0.7)]

    X_train = X_newdata[:int(len(X_newdata)*0.7)+1]
    X_test = X_newdata[int(len(X_newdata)*0.7):]
    y_train = Y_data.iloc[:,i][:int(len(Y_data.iloc[:,i])*0.7)+1]
    y_test = Y_data.iloc[:,i][int(len(Y_data.iloc[:,i])*0.7):]

        
    # 各个模型的交叉验证
    if i == 0:
        models = {
            "线性回归":LinearRegression(),
            "决策树":DecisionTreeRegressor(),
            "SVM":SVR(),
            "Lasso":Lasso(),
            "XGBoost":XGBRegressor(),
            "AdaBoost":AdaBoostRegressor()}

        scores_dict = {}
        for name,model in models.items():
            cv = ShuffleSplit(n_splits=10, test_size=.3, random_state=0) 
            scores_CV = cross_val_score(model, X_newdata, Y_data.iloc[:,i], cv=cv)  
            scores_dict[name] = scores_CV  # 保存交叉验证评分结果

        # 结果
        df_scoresCV = pd.DataFrame(scores_dict)
        df_scoresCV["变量"]=Y_data.columns[i]
        if i== 0:
            scorecv_output = df_scoresCV
        else:
            scorecv_output = pd.concat([scorecv_output,df_scoresCV])


        # 可视化绘制箱线图
        score_plotdata = pd.DataFrame(scores_dict).melt()

        ax=fig.add_subplot(1,1,i+1)
        sns.boxplot(x="variable", y="value",data=score_plotdata)
        sns.swarmplot(x="variable", y="value", data=score_plotdata, color=".25")
    #     ax.set_xticks(rotation=90)
        ax.set_xlabel(Y_data.columns[i])
        ax.set_ylabel("Score")


    if i==0:
        """线性回归"""
        lr = LinearRegression(fit_intercept=True,n_jobs=10).fit(X_train,y_train)
        res_models.append(lr)
        lr_coef = np.round(lr.coef_,3)
        print("{}线性回归系数：\n".format(Y_data.columns[i]),lr_coef)
    
        
        print('Training set score:{:.10f}\n'.format(lr.score(X_train,y_train)))
        print('Test set score:{:.10f}\n'.format(lr.score(X_test,y_test)))
        print("Training set MSE:{:.2f}\n".format(mean_squared_error(y_train,lr.predict(X_train))))
        print("Test set MSE:{:.2f}\n".format(mean_squared_error(y_test,lr.predict(X_test))))
    
    
scorecv_output.to_csv("./output/交叉验证评分结果.csv",index=False,header=True,encoding = 'utf_8_sig')
plt.savefig("./output/交叉验证评分结果箱线图.png",dpi=500, bbox_inches = "tight")
plt.savefig("./output/交叉验证评分结果箱线图.pdf",dpi=500, bbox_inches = "tight")

plt.show()
```
